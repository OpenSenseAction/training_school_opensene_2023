# 2. Quick intro to Jupyterlab

### 2.1 The two input modes

The IPython Notebook has **two different keyboard input modes**. 

 * **Edit mode** (press `Return` to enable) allows you to type code/text into a cell and is indicated by a green cell border. 

 * **Command mode** (press `Esc` to enable) binds the keyboard to notebook level actions and is indicated by a grey cell border.

### 2.2 Creating and executing cells

Create a cell **in command mode** by pressing 
   * **`a`** for above the current cell
   * **`b`** for below the current cell   

Execute a cell **in edit mode** by pressing
   * **`Ctrl + Return`** and stay in the current cell
   * **`Shift + Return`** and automatically go to the next cell
   * **`Alt + Return`** and automatically create a new cell below

### 2.3 Two different cell types
